```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
h2_jw=[21,61,57]
h2_jw_small=[10,34,24]
h2_bk=[4,13,26]
h2_bk_small=[10,34,45]
```

```python
barWidth = 0.1
```

```python
# Set position of bar on X axis
r1 = np.arange(len(h2_jw))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]
r4 = [x + barWidth for x in r3]
```

```python
# Make the plot
plt.bar(r1, h2_jw, color='#C70039', width=barWidth, edgecolor='white', label='Jordan-Wigner ')
plt.bar(r2, h2_jw_small, color='#FFC300', width=barWidth, edgecolor='white', label='Small Jordan-Wigner')
plt.bar(r3, h2_bk, color='#ff5733' , width=barWidth, edgecolor='white', label='Brayvi-Kitaev')
plt.bar(r4, h2_bk_small, color= '#DAF7A6',  width=barWidth, edgecolor='white', label='Small Bravyi-Kitaev')
# Add xticks on the middle of the group bars
plt.ylabel("Function Evaluasi")
plt.xticks([r + barWidth for r in range(len(h2_jw))], ['SLSQP', 'Nelder-Mead', 'Powell'])
 
# Create legend & Show graphic
plt.legend()
plt.grid()
plt.show()
```
